```python
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = df["YEAR"]
y = df["ANNUAL"]
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=x,y=y,data=df)
```

```python
x.shape
```

```python
x = x.reshape(117,1)
```

```python
x.shape
```

```python
type(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25)

print(f"x Training dataset: {x_train.shape}")
print(f"y Training dataset: {y_train.shape}")
print(f"x test dataset: {x_test.shape}")
print(f"y test dataset: {y_test.shape}")
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MSE:  {mean_squared_error(y_test,y_pred)}")
print(f"MAE:  {mean_absolute_error(y_test,y_pred)}")
print(f"R-Sqaure :  {r2_score(y_test,y_pred)}")
```

```python
plt.scatter(x_train, y_train, color='green')
plt.plot(x_test, y_pred, color='black', linewidth=4)
plt.title("Temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

```python

import numpy as np
year_to_predict = 2028
year_to_predict = np.array(year_to_predict).reshape(1, -1)
predicted_temp = model.predict(year_to_predict)
print(f"Predicted temperature for year {year_to_predict[0][0]}: {predicted_temp[0]}")
```
